```python
import pandas as pd
import matplotlib 
%matplotlib notebook 
import matplotlib.pyplot as plt 
import os
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

df_iris = pd.read_csv("iris.data", header=None, names=columns)
```

```python
print(df_iris.shape)

df_iris.sample(n=10, replace=False, random_state = 1)
```

```python
X = df_iris[['petal_length', 'petal_width']]
y = df_iris['species']

print("Num of Examples: " + str(X.shape[0]))
print("Num of Features: " + str(X.shape[1]))

print("\nFeatures: \n")
print(X)

print("\nLabels: \n")
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
# initialize model obj 
model = KNeighborsClassifier(n_neighbors = 3)
# train/fit
model.fit(X_train, y_train)
# predict
prediction = model.predict(X_test)
print(prediction)
```

```python
score = accuracy_score(y_test, prediction)
print("Accuracy score of Model: " + str(score))
```
